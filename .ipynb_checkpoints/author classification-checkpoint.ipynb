```python
import pandas as pd
import os
```

```python
os.listdir()
```

```python
df = pd.read_csv('medium_articles.csv')
```

```python
df.head()
```

```python
ls = df.authors.unique().tolist()
```

```python
len(ls)
```

```python
def prepro_author(text):
    if(text[0] == '[' and text[1] == ']' and len(text)==2):
        return ("NONE")
    else:
        return text[2:len(text)-2]
```

```python
df.authors = df.authors.apply(lambda x: prepro_author(x))
```

```python
df.head()
```

```python
df.text = df.text.str.replace("\n"," ")
```

```python
df.head()
```

```python
tmp_new = df.loc[df.authors != 'NONE']
```

```python
tmp_new
```

```python
tmp_new.authors.value_counts()
```

```python
ar_df = tmp_new.loc[(tmp_new.authors == 'Are You Syrious') | (tmp_new.authors == 'John Au-Yeung') | (tmp_new.authors == 'The Good Men Project') | (tmp_new.authors == 'J.D. Harms')] 
```

```python
ar_df_title = ar_df[['title','authors']]
ar_df_text = ar_df[['text','authors']]
```

```python
ar_df_title
```

```python
ar_df_text
```

```python
ls = ['Are You Syrious','John Au-Yeung','The Good Men Project','J.D. Harms']
```

```python
ar_df_title.authors = ar_df_title.authors.apply(lambda x: ls.index(x))
```

```python
ar_df_text.authors = ar_df_text.authors.apply(lambda x: ls.index(x))
```

```python
ar_df_title
```

```python
ar_df_text
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=10)
```

```python
metric_dict_title = {
    '0': {'precision': [], 'recall': [], 'f1-score': []},
    '1': {'precision': [], 'recall': [], 'f1-score': []},
    '2': {'precision': [], 'recall': [], 'f1-score': []},
    '3': {'precision': [], 'recall': [], 'f1-score': []},
    'accuracy':[]
}
metric_dict_text = {
    '0': {'precision': [], 'recall': [], 'f1-score': []},
    '1': {'precision': [], 'recall': [], 'f1-score': []},
    '2': {'precision': [], 'recall': [], 'f1-score': []},
    '3': {'precision': [], 'recall': [], 'f1-score': []},
    'accuracy':[]
}
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
i=1
for train_index, test_index in skf.split(ar_df_title.title, ar_df_title.authors):
    X_train, X_test = ar_df_title.title.iloc[train_index], ar_df_title.title.iloc[test_index]
    y_train, y_test = ar_df_title.authors.iloc[train_index], ar_df_title.authors.iloc[test_index]
    vectorizer_title = TfidfVectorizer()
    X_vector_title = vectorizer_title.fit_transform(X_train)
    X_vector_test_title = vectorizer_title.transform(X_test)
    from sklearn.svm import LinearSVC
    clf_svm_title = LinearSVC(random_state=0)
    clf_svm_title.fit(X_vector_title,y_train)
    from sklearn.metrics import classification_report
    y_pred = clf_svm_title.predict(X_vector_test_title)
    print(f"fold: {i}")
    cl_dict = classification_report(y_test, y_pred,output_dict=True)
#     append value to metric_dict_title
    for i in range(0,4):    
        metric_dict_title[str(i)]['precision'].append(cl_dict[str(i)]['precision'])
        metric_dict_title[str(i)]['recall'].append(cl_dict[str(i)]['recall'])
        metric_dict_title[str(i)]['f1-score'].append(cl_dict[str(i)]['f1-score'])
    metric_dict_title['accuracy'].append(cl_dict['accuracy'])
    i+=1
```

```python
metric_dict_title
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
i=1
for train_index, test_index in skf.split(ar_df_text.text, ar_df_text.authors):
    X_train, X_test = ar_df_text.text.iloc[train_index], ar_df_text.text.iloc[test_index]
    y_train, y_test = ar_df_text.authors.iloc[train_index], ar_df_text.authors.iloc[test_index]
    vectorizer = TfidfVectorizer()
    X_vector = vectorizer.fit_transform(X_train)
    X_vector_test = vectorizer.transform(X_test)
    from sklearn.svm import LinearSVC
    clf_svm = LinearSVC(random_state=0)
    clf_svm.fit(X_vector,y_train)
    from sklearn.metrics import classification_report
    y_pred = clf_svm.predict(X_vector_test)
    cl_dict = classification_report(y_test, y_pred,output_dict=True)
#     append value to metric_dict_title
    for i in range(0,4):    
        metric_dict_title[str(i)]['precision'].append(cl_dict[str(i)]['precision'])
        metric_dict_title[str(i)]['recall'].append(cl_dict[str(i)]['recall'])
        metric_dict_title[str(i)]['f1-score'].append(cl_dict[str(i)]['f1-score'])
    metric_dict_title['accuracy'].append(cl_dict['accuracy'])
    i+=1
```

```python
from sklearn.model_selection import train_test_split
X_train_title, X_test_title, y_train_title, y_test_title = train_test_split(ar_df_title.title, ar_df_title.authors, test_size=0.2, random_state=42, stratify=ar_df_title.authors)
```

```python
from sklearn.model_selection import train_test_split
X_train_text, X_test_text, y_train_text, y_test_text = train_test_split(ar_df_text.text, ar_df_text.authors, test_size=0.2, random_state=42, stratify=ar_df_text.authors)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_title = TfidfVectorizer()
vectorizer_text = TfidfVectorizer()
X_vector_title = vectorizer_title.fit_transform(X_train_title)
X_vector_text = vectorizer_text.fit_transform(X_train_text)
```

```python
X_vector_text.shape
```

```python
X_vector_test_title = vectorizer_title.transform(X_test_title)
X_vector_test_text = vectorizer_text.transform(X_test_text)
```

```python
X_vector_test_text.shape
```

```python
from sklearn.svm import LinearSVC
clf_svm_title = LinearSVC(random_state=0)
clf_svm_title.fit(X_vector_title,y_train_title)
```

```python
y_pred_title = clf_svm_title.predict(X_vector_test_title)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_title, y_pred_title))
```

```python
clf_svm_text = LinearSVC(random_state=0)
clf_svm_text.fit(X_vector_text,y_train_text)
```

```python
y_pred_text = clf_svm_text.predict(X_vector_test_text)
print(classification_report(y_test_text, y_pred_text))
```
