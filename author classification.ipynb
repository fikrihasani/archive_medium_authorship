```python
def prepro_author(text):
    from ast import literal_eval
    ls = literal_eval(text)
    return ls[0] if len(ls) > 0 else ''
```

```python
df.authors = df.authors.apply(lambda x: prepro_author(x))
```

```python
df.head()
```

```python
df.text = df.text.str.replace("\n"," ")
```

```python
df.head()
```

```python
tmp_new = df.loc[df.authors != '']
```

```python
tmp_new
```

```python
ll = list(tmp_new.authors.value_counts().to_dict().items())
```

```python
dt = [x for x in ll if x[1] >= 100]
```

```python
dt
```

```python
ar_df = pd.DataFrame()
for items in dt:
     ar_df = ar_df.append(tmp_new.loc[tmp_new.authors == items[0]],ignore_index=True)
```

```python
ar_df.head()
```

```python
ar_df.text = ar_df.text.str.lower()
ar_df.title = ar_df.title.str.lower()
```

```python
ar_df.text = ar_df.text.str.replace('[^\w\s]',' ')
ar_df.title = ar_df.title.str.replace('[^\w\s]',' ')
```

```python
ar_df.head()
```

```python
ar_df_title = ar_df[['title','authors']]
ar_df_text = ar_df[['text','authors']]
```

```python
ar_df_title
```

```python
ar_df_text
```

```python
def getAuthorIdx(author):
    for i,items in enumerate(dt):
        if items[0] == author:
            return i
```

```python
ar_df_title.authors = ar_df_title.authors.apply(lambda x: getAuthorIdx(x))
```

```python
ar_df_text.authors = ar_df_text.authors.apply(lambda x: getAuthorIdx(x))
```

# 1 - Gram

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=10)
```

```python
def init_dict(dc, df ):
    for i in range(len(df.authors.unique())):
        dc[i] = {'precision': [], 'recall': [], 'f1-score': []}
    dc['accuracy'] = []
```

```python
met = {}
init_dict(met,ar_df_title)
```

```python
met
```

## Init dictionary

```python
metric_dict_title_svm = {}
init_dict(metric_dict_title_svm, ar_df_title)

metric_dict_title_knn = {}
init_dict(metric_dict_title_knn, ar_df_title)

metric_dict_title_randomforest = {}
init_dict(metric_dict_title_randomforest, ar_df_title)

metric_dict_title_naivebayes = {}
init_dict(metric_dict_title_naivebayes, ar_df_title)
```

```python
metric_dict_title_knn
```

```python
metric_dict_text_svm = {}
init_dict(metric_dict_text_svm, ar_df_title)

metric_dict_text_knn = {}
init_dict(metric_dict_text_knn, ar_df_title)

metric_dict_text_randomforest = {}
init_dict(metric_dict_text_randomforest, ar_df_title)

metric_dict_text_naivebayes = {}
init_dict(metric_dict_text_naivebayes, ar_df_title)
```

```python
def calc_dict_metric(dc, df, matrix):
    for i in range(len(df.authors.unique())):    
        dc[i]['precision'].append(matrix[str(i)]['precision'])
        dc[i]['recall'].append(matrix[str(i)]['recall'])
        dc[i]['f1-score'].append(matrix[str(i)]['f1-score'])
    dc['accuracy'].append(matrix['accuracy'])
```

## Train title model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report

j=1
for train_index, test_index in skf.split(ar_df_title.title, ar_df_title.authors):
    X_train, X_test = ar_df_title.title.iloc[train_index], ar_df_title.title.iloc[test_index]
    y_train, y_test = ar_df_title.authors.iloc[train_index], ar_df_title.authors.iloc[test_index]
    vectorizer_title = TfidfVectorizer()
    X_vector_title = vectorizer_title.fit_transform(X_train)
    X_vector_test_title = vectorizer_title.transform(X_test)
    print(j)
    print("SVM Classifier")
    # Linear SVM
    from sklearn.svm import LinearSVC
    clf_svm_title = LinearSVC(random_state=0)
    clf_svm_title.fit(X_vector_title,y_train)
    y_pred = clf_svm_title.predict(X_vector_test_title)
    cl_dict_svm = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("KNN Classifier")
    # KNN
    from sklearn.neighbors import KNeighborsClassifier
    clf_knn_title = KNeighborsClassifier(n_neighbors=5)
    clf_knn_title.fit(X_vector_title,y_train)
    y_pred = clf_knn_title.predict(X_vector_test_title)
    cl_dict_knn = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Random Forest Classifier")
    # Random forest
    from sklearn.ensemble import RandomForestClassifier
    clf_rf_title = RandomForestClassifier()
    clf_rf_title.fit(X_vector_title,y_train)
    y_pred = clf_rf_title.predict(X_vector_test_title)
    cl_dict_rf = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Naive Bayes Classifier")
    from sklearn.naive_bayes import MultinomialNB
    clf_nb_title = RandomForestClassifier()
    clf_nb_title.fit(X_vector_title,y_train)
    y_pred = clf_nb_title.predict(X_vector_test_title)
    cl_dict_nb = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    
    calc_dict_metric(metric_dict_title_svm, ar_df_title, cl_dict_svm)
    calc_dict_metric(metric_dict_title_knn, ar_df_title, cl_dict_knn)
    calc_dict_metric(metric_dict_title_randomforest, ar_df_title, cl_dict_rf)
    calc_dict_metric(metric_dict_title_naivebayes, ar_df_title, cl_dict_nb)
    
    j+=1
```

## train text model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
j=1
for train_index, test_index in skf.split(ar_df_text.text, ar_df_text.authors):
    X_train, X_test = ar_df_text.text.iloc[train_index], ar_df_text.text.iloc[test_index]
    y_train, y_test = ar_df_text.authors.iloc[train_index], ar_df_text.authors.iloc[test_index]
    vectorizer = TfidfVectorizer()
    X_vector = vectorizer.fit_transform(X_train)
    X_vector_test = vectorizer.transform(X_test)

    print(f"fold: {j}")
    print("SVM Classifier")
    # Linear SVM
    from sklearn.svm import LinearSVC
    clf_svm_text = LinearSVC(random_state=0)
    clf_svm_text.fit(X_vector,y_train)
    y_pred = clf_svm_text.predict(X_vector_test)
    cl_dict_svm = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("KNN Classifier")
    # KNN
    from sklearn.neighbors import KNeighborsClassifier
    clf_knn_text = KNeighborsClassifier()
    clf_knn_text.fit(X_vector,y_train)
    y_pred = clf_knn_text.predict(X_vector_test)
    cl_dict_knn = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Random Forest Classifier")
    # Random forest
    from sklearn.ensemble import RandomForestClassifier
    clf_rf_text = RandomForestClassifier()
    clf_rf_text.fit(X_vector,y_train)
    y_pred = clf_rf_text.predict(X_vector_test)
    cl_dict_rf = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Naive Bayes Classifier")
    from sklearn.naive_bayes import MultinomialNB
    clf_nb_text = RandomForestClassifier()
    clf_nb_text.fit(X_vector,y_train)
    y_pred = clf_nb_text.predict(X_vector_test)
    cl_dict_nb = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    
    calc_dict_metric(metric_dict_text_svm, ar_df_title, cl_dict_svm)
    calc_dict_metric(metric_dict_text_knn, ar_df_title, cl_dict_knn)
    calc_dict_metric(metric_dict_text_randomforest, ar_df_title, cl_dict_rf)
    calc_dict_metric(metric_dict_text_naivebayes, ar_df_title, cl_dict_nb)
    print()
    j+=1
```

```python
def calc_avg_met(dc, df, met_dc):
    for i in range(len(df.authors.unique())):
        dc[i] = {}
        dc[i]['precision'] = sum(met_dc[i]['precision'])/len(met_dc[i]['precision'])
        dc[i]['recall'] = sum(met_dc[i]['recall'])/len(met_dc[i]['recall'])
        dc[i]['f1-score'] = sum(met_dc[i]['f1-score'])/len(met_dc[i]['f1-score'])
    dc['accuracy'] = sum(met_dc['accuracy'])/len(met_dc['accuracy'])
```

## Calculate average of k-fold

```python
avg_title_svm = {}
calc_avg_met(avg_title_svm, ar_df_title, metric_dict_title_svm)

avg_title_rf = {}
calc_avg_met(avg_title_rf, ar_df_title, metric_dict_title_randomforest)

avg_title_knn = {}
calc_avg_met(avg_title_knn, ar_df_title, metric_dict_title_knn)

avg_title_nb = {}
calc_avg_met(avg_title_nb, ar_df_title, metric_dict_title_naivebayes)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
j=1
for train_index, test_index in skf.split(ar_df_text.text, ar_df_text.authors):
    X_train, X_test = ar_df_text.text.iloc[train_index], ar_df_text.text.iloc[test_index]
    y_train, y_test = ar_df_text.authors.iloc[train_index], ar_df_text.authors.iloc[test_index]
    vectorizer = TfidfVectorizer(ngram_range=(3,3))
    X_vector = vectorizer.fit_transform(X_train)
    X_vector_test = vectorizer.transform(X_test)

    print(f"fold: {j}")
    print("SVM Classifier")
    # Linear SVM
    from sklearn.svm import LinearSVC
    clf_svm_text = LinearSVC(random_state=0)
    clf_svm_text.fit(X_vector,y_train)
    y_pred = clf_svm_text.predict(X_vector_test)
    cl_dict_svm = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("KNN Classifier")
    # KNN
    from sklearn.neighbors import KNeighborsClassifier
    clf_knn_text = KNeighborsClassifier()
    clf_knn_text.fit(X_vector,y_train)
    y_pred = clf_knn_text.predict(X_vector_test)
    cl_dict_knn = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Random Forest Classifier")
    # Random forest
    from sklearn.ensemble import RandomForestClassifier
    clf_rf_text = RandomForestClassifier()
    clf_rf_text.fit(X_vector,y_train)
    y_pred = clf_rf_text.predict(X_vector_test)
    cl_dict_rf = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    print("Naive Bayes Classifier")
    from sklearn.naive_bayes import MultinomialNB
    clf_nb_text = RandomForestClassifier()
    clf_nb_text.fit(X_vector,y_train)
    y_pred = clf_nb_text.predict(X_vector_test)
    cl_dict_nb = classification_report(y_test, y_pred,output_dict=True, zero_division=0)

    
    calc_dict_metric(metric_dict_text_3_svm, ar_df_title, cl_dict_svm)
    calc_dict_metric(metric_dict_text_3_knn, ar_df_title, cl_dict_knn)
    calc_dict_metric(metric_dict_text_3_randomforest, ar_df_title, cl_dict_rf)
    calc_dict_metric(metric_dict_text_3_naivebayes, ar_df_title, cl_dict_nb)

    print()
    j+=1
```
